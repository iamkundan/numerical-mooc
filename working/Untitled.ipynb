Comuting 1 d advection equation using Forward in Time and forward in space method (FDM)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

solve 1 d linear convection equation with a *square wave* initial condition :

$$
\begin{equation}
u(x,0)= \begin{cases}2 & \text {where }0.5\leq x \leq1 \\, 1 & \text {everywhere else in } (0,2)
\end{cases}
\end {equation}
$$

```python
# define variables

nx=41
L=2.0
dx=L/(nx-1)
nt =20
dt =0.02
c =1.0

x=np.linspace(0.0,L,num=nx)
x
```

```python
u0=np.ones(nx)
u0.shape
mask=np.where(np.logical_and(x>=0.5, x<=1.0))
print (mask)
```

```python
u0[mask]=2.0
print (u0)
```

```python
plt.plot(x,u0)
```

```python
u=u0.copy()
for n in range(1,nt):
    #print("Iteration Time no.:",n\n)
    print(n, un)
    un=u.copy()
    for i in range(1,nx):
        print(i)
        #print("Iteration space no.:",i\n)
        u[i]=un[i]-c*dt/dx*(un[i]-un[i-1])
```

```python
plt.figure(figsize=(4.0,4.0))
plt.xlabel('x')
plt.ylabel('u')
plt.grid()
plt.plot(x,u0, label='Initial', color='C0', linestyle='--', linewidth=2)
plt.plot(x,u,label='nt={}'.format(nt), color='C1', linestyle='-', linewidth=2)
plt.legend()
plt.xlim(0.0,L)
plt.ylim(0.0, 2.5)
```
